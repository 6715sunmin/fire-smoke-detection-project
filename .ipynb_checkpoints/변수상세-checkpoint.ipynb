```python
data_group = data.groupby('Fire Alarm').mean().iloc[:, :]
data_group
```

```python
for i in range(2):
    plt.subplot(1,2,i+1)
    plt.imshow(np.array(data_group.iloc[i]).reshape(13,1))
    plt.title(f'Fire Alarm : {i}')
    plt.xlim(-1,1)
    plt.xticks(fontsize=10)
```

```python
fire0 = data[data['Fire Alarm']==0]
fire1 = data[data['Fire Alarm']==1]
```

```python
sns.histplot(data=fire0, x='Temperature[C]', element='poly', stat='percent', alpha=0.3, label='Fire Alarm 0')
sns.histplot(data=fire1, x='Temperature[C]', element='poly', stat='percent', alpha=0.3, label='Fire Alarm 1')
plt.legend()
```

```python
sns.histplot(data=fire0, x='Humidity[%]', element='poly', stat='percent', alpha=0.3, label='Fire Alarm 0')
sns.histplot(data=fire1, x='Humidity[%]', element='poly', stat='percent', alpha=0.3, label='Fire Alarm 1')
plt.legend()
```

```python
sns.histplot(data=fire0, x='Raw H2', element='poly', stat='percent', alpha=0.3, label='Fire Alarm 0')
sns.histplot(data=fire1, x='Raw H2', element='poly', stat='percent', alpha=0.3, label='Fire Alarm 1')
plt.legend()
```

```python
sns.histplot(data=fire0, x='Pressure[hPa]', element='poly', stat='percent', alpha=0.3, label='Fire Alarm 0')
sns.histplot(data=fire1, x='Pressure[hPa]', element='poly', stat='percent', alpha=0.3, label='Fire Alarm 1')
plt.legend()
```
